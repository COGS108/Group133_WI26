## Research Question

Were higher per-capita $CO_2$ emissions associated with greater climate-related disaster impacts (in the form of higher mortality and reduced GDP growth) for lower-income countries than for higher-income countries between the years 2005-2023?

## Background and Prior Work

Global warming is largely caused by the accumulation of human-emitted greenhouse gases (GHG), namely, carbon dioxide ($CO_2$), which accounts for about 66% of human-emitted GHGs. <a name="cite_note-1"></a><sup>1</sup> Some sources of emission include burning fossil fuels for energy, transportation, and manufacturing. Global datasets, such as those maintained by Our World in Data, provide long-term measurements of $CO_2$ emissions, as well as their trends across countries. They show clear differences between high- and low-income nations in per-capita emission levels. <a name="cite_note-2"></a><sup>2</sup> In addition, the rise in $CO_2$ emissions result in detrimental climate change, with shifting climate patterns increasing the severity and frequency of extreme weather events such as heatwaves, floods, and storms. <a name="cite_note-3"></a><sup>3</sup> The differences between high- and low-income nations in per-capita emission levels, as well as the climate-disaster impacts experienced, raises questions about climate inequality.
\
\
There has been growing research and discussion on climate inequality, which explores the ways in which climate change may disproportionately impact different groups of people. International disaster records, such as those by EM-DAT, show that disaster-related mortality rates vary over time and across regions, with certain countries experiencing significantly higher mortality rates per 100,000 people than others. <a name="cite_note-4"></a><sup>4</sup> Despite contributing far less to GHG emissions, lower-income countries and vulnerable populations typically experience larger relative burdens of climate-driven losses, compared to wealthier nations and privileged populations. <a name="cite_note-5"></a><sup>5</sup>
\
\
We can examine these climate-related disaster impacts in the form of higher mortality and economic strain. Research from global monitoring organizations highlight that economic losses are especially exacerbated in regions with weaker infrastructure and lower capacity for disaster response. <a name="cite_note-6"></a><sup>6</sup> For example, climate-related harms can greatly reduce GDP growth in poorer regions and slow overall economic development, further contributing to the wider disparity in wealth and well-being between countries. <a name="cite_note-7"></a><sup>7</sup> Finally, multiple analyses and systematic reviews show that inequalities in capacity and access to resources can compound the harmful effects of climate change, thus worsening the outcomes for these disadvantaged groups both within and across countries. <a name="cite_note-8"></a><sup>8</sup>
\
\
These findings suggest that combining measures of $CO_2$ emissions and disaster impacts, in the form of higher mortality and reduced GDP per capita, can show us different countries' climate contribution versus vulnerability. By utilizing publicly available datasets on per-capita $CO_2$ emissions, mortality rates, and GDP per capita, we aim to quantify the relationship between human activity and climate outcomes across low- and high-income countries. This approach builds on prior work in climate change science and economic inequality research while applying data science and analysis practices and methods covered in this course.

## Hypothesis


We hypothesize that when countries are grouped by income level, the association between $CO_2$ emissions and disaster impacts will differ in that:

- Lower-income countries will experiene a greater ratio of climate-related disaster impacts to per-capita $CO_2$ emissions. 

- Higher-income countries will experience a lower ratio of climate-related disaster impacts to per-capita $CO_2$ emissions. 


*Note: Climate-related disaster impacts is defined by both increased mortality rate and decreased GDP per capita.*
