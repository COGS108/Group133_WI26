# Research Question

Were higher per-capita $CO_2$ emissions associated with greater climate-related disaster impacts (in the form of higher mortality and reduced GDP growth) for lower-income countries than for higher-income countries between the years 2005-2023?

---

## Background and Prior Work

Global warming is largely caused by the accumulation of human-emitted greenhouse gases (GHG), namely, carbon dioxide ($CO_2$), which accounts for about 66% of human-emitted GHGs. <a name="cite_note-1"></a><sup>1</sup> Some sources of emission include burning fossil fuels for energy, transportation, and manufacturing. Global datasets, such as those maintained by Our World in Data, provide long-term measurements of $CO_2$ emissions, as well as their trends across countries. They show clear differences between high- and low-income nations in per-capita emission levels. <a name="cite_note-2"></a><sup>2</sup> In addition, the rise in $CO_2$ emissions result in detrimental climate change, with shifting climate patterns increasing the severity and frequency of extreme weather events such as heatwaves, floods, and storms. <a name="cite_note-3"></a><sup>3</sup> The differences between high- and low-income nations in per-capita emission levels, as well as the climate-disaster impacts experienced, raises questions about climate inequality.
\
\
There has been growing research and discussion on climate inequality, which explores the ways in which climate change may disproportionately impact different groups of people. International disaster records, such as those by EM-DAT, show that disaster-related mortality rates vary over time and across regions, with certain countries experiencing significantly higher mortality rates per 100,000 people than others. <a name="cite_note-4"></a><sup>4</sup> Despite contributing far less to GHG emissions, lower-income countries and vulnerable populations typically experience larger relative burdens of climate-driven losses, compared to wealthier nations and privileged populations. <a name="cite_note-5"></a><sup>5</sup>
\
\
We can examine these climate-related disaster impacts in the form of higher mortality and economic strain. Research from global monitoring organizations highlight that economic losses are especially exacerbated in regions with weaker infrastructure and lower capacity for disaster response. <a name="cite_note-6"></a><sup>6</sup> For example, climate-related harms can greatly reduce GDP growth in poorer regions and slow overall economic development, further contributing to the wider disparity in wealth and well-being between countries. <a name="cite_note-7"></a><sup>7</sup> Finally, multiple analyses and systematic reviews show that inequalities in capacity and access to resources can compound the harmful effects of climate change, thus worsening the outcomes for these disadvantaged groups both within and across countries. <a name="cite_note-8"></a><sup>8</sup>
\
\
These findings suggest that combining measures of $CO_2$ emissions and disaster impacts, in the form of higher mortality and reduced GDP per capita, can show us different countries' climate contribution versus vulnerability. By utilizing publicly available datasets on per-capita $CO_2$ emissions, mortality rates, and GDP per capita, we aim to quantify the relationship between human activity and climate outcomes across low- and high-income countries. This approach builds on prior work in climate change science and economic inequality research while applying data science and analysis practices and methods covered in this course.

# Hypothesis


We hypothesize that when countries are grouped by income level, the association between $CO_2$ emissions and disaster impacts will differ in that:

- Lower-income countries will experiene a greater ratio of climate-related disaster impacts to per-capita $CO_2$ emissions. 

- Higher-income countries will experience a lower ratio of climate-related disaster impacts to per-capita $CO_2$ emissions. 


*Note: Climate-related disaster impacts is defined by both increased mortality rate and decreased GDP per capita.*

### Load Datasets #1, #2, #3

```python
import pandas as pd

#shared time frame of datasets
START_YEAR = 2005
END_YEAR = 2023

#OWID URLs with short column names 
co2_url = "https://ourworldindata.org/grapher/co-emissions-per-capita.csv?v=1&csvType=full&useColumnShortNames=true"
disasters_url = "https://ourworldindata.org/grapher/deaths-and-missing-persons-due-to-natural-disasters.csv?v=1&csvType=full&useColumnShortNames=true"
gdp_url = "https://ourworldindata.org/grapher/gdp-per-capita-worldbank.csv?v=1&csvType=full&useColumnShortNames=true"

#load dataset, from website, storage option specified
co2 = pd.read_csv(co2_url, storage_options={'User-Agent': 'Our World In Data data fetch/1.0'})
disasters = pd.read_csv(disasters_url, storage_options={'User-Agent': 'Our World In Data data fetch/1.0'})
gdp = pd.read_csv(gdp_url, storage_options={'User-Agent': 'Our World In Data data fetch/1.0'})

print('Loaded datasets:')
print('CO2 shape:', co2.shape)
print('Disasters shape:', disasters.shape)
print('GDP shape:', gdp.shape)
```

### Filter to 2005-2023 and confine to countries

```python
#restrict to shared time frame, will use .copy() for extra safety
co2 = co2[(co2["year"] >= START_YEAR) & (co2["year"] <= END_YEAR)].copy()
disasters = disasters[(disasters["year"] >= START_YEAR) & (disasters["year"] <= END_YEAR)].copy()
gdp = gdp[(gdp["year"] >= START_YEAR) & (gdp["year"] <= END_YEAR)].copy()

#remove aggregate regions such as 'Asia', 'World', etc to only look at countries - keep rows with ISO codes
co2 = co2[co2["code"].notna()].copy()
disasters = disasters[disasters["code"].notna()].copy()
gdp = gdp[gdp["code"].notna()]

print('After filtering:')
print('CO2 Shape:', co2.shape)
print('Disasters shape:', disasters.shape)
print('GDP shape:', gdp.shape)
```

### Inspect and rename key variables for readability

```python
#select and rename key variables after previously analyzing column names with df.columns
co2_clean=co2[['entity', 'code', 'year', 'emissions_total_per_capita']].rename(columns = {'emissions_total_per_capita': 'co2_tons_per_capita'})
disasters_clean = disasters[['entity', 'code', 'year', '_13_1_1__vc_dsr_mtmp']].rename(columns={'_13_1_1__vc_dsr_mtmp': 'disaster_deaths_missing_per_100k'})
gdp_clean = gdp[['entity', 'code', 'year', 'ny_gdp_pcap_pp_kd', 'owid_region']].rename(columns = {'ny_gdp_pcap_pp_kd': 'gdp_per_capita_ppp_const'})

print('CO2 clean columns:', co2_clean.columns.tolist())
print('Disasters clean columns:', disasters_clean.columns.tolist())
print('GDP clean columns:', gdp_clean.columns.tolist())
```

### Merge into single, final wrangled dataframe

```python
#merge using ISO and year
df = co2_clean.merge(disasters_clean[['code', 'year', 'disaster_deaths_missing_per_100k']], on = ['code', 'year'], how = 'inner')
df = df.merge(gdp_clean[['code', 'year', 'gdp_per_capita_ppp_const', 'owid_region']], on = ['code', 'year'], how = 'inner')

#drop any missing values in key variables
df = df.dropna(subset=['co2_tons_per_capita', 'disaster_deaths_missing_per_100k', 'gdp_per_capita_ppp_const']).copy()

#remove duplicate country year rows in case
df = df.drop_duplicates(subset = ['code', 'year'])

print('Final merged dataset shape:', df.shape) 
print('Countries:', df['entity'].nunique())
print('Years:', df['year'].min(), 'to', df['year'].max())
df.head()
```

Understanding Per-Capita CO₂ Emissions Across Income Levels
\
\
To understand climate inequality, we will first look at how economic development relates to carbon emissions. This section will explore the fundamental relationship between a country's wealth (measured by GDP per capita) and its per-capita CO₂ emissions. As previously stated, we expect higher-income countries to emit more CO₂.
\
\
By examining this wealth-emissions relationship, we are establishing the foundation necessary to further explore  our hypothesis that lower-income countries  experience disproportionately higher disaster impacts relative to their emissions contributions.

```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv("data/02-processed/climate_inequality_2005_2023.csv")

fig, ax = plt.subplots(figsize=(12,7)) 
sns.set_theme(style="whitegrid") 

## Create scatter plot 
sns.scatterplot(
    data=df, 
    x='gdp_per_capita_ppp_const', 
    y='co2_tons_per_capita', 
    hue='owid_region', 
    alpha=0.6, 
    edgecolor='w', 
    palette='viridis',
    ax=ax
)

## Apply log scales 
ax.set_xscale("log") 
ax.set_yscale("log")

plt.title('The Wealth-Emissions Gap (2005-2023)', fontsize=16, fontweight='bold', pad=20) 
plt.xlabel('GDP per Capita (PPP, Constant 2017 Intl $)', fontsize=12) 
plt.ylabel('CO2 Emissions (Tons per Capita)', fontsize=12) 
plt.legend(title='Region', bbox_to_anchor=(1.05, 1), loc='upper left') 

sns.despine() 
plt.tight_layout() 
plt.show()
```

#### Observations from the Wealth-Emissions Relationship Scatterplot

The scatter plot above reveals to us a striking positive relationship between GDP per capita and per-capita CO₂ emissions, specifically between 2005-2023. 

Lets consider several key patterns that emerge:

**Strong Regional Clustering:** Nations cluster by region and income level, with European and developed Asian economies showing both high wealth and high emissions, while African and least-developed nations cluster in the lower-left, indicating both lower GDP and lower per-capita emissions.

**Emissions as a Development Marker:** The log-linear relationship suggests to us that as economies develop and wealth increases, per-capita emissions tend to rise. This is likely due to increased energy consumption, industrial activity, and transportation.

**Outliers of Interest:** A few nations demonstrate different patterns—some wealthy nations, as they show  relatively lower emissions (possibly due to renewable energy adoption), while others show high emissions despite moderate wealth (often resource-dependent economies).

This wealth-emissions relationship is a foundation that is crucial for understanding our hypothesis: if wealthier nations both emit more AND experience fewer disaster impacts per unit of emissions, while poorer nations emit less but suffer greater impacts, this would reveal profound climate inequality.

#### Summary: What Section 1 Data Tells Us

Section 1 demonstrates to us that wealth and emissions are intimately connected. However, emissions alone do not tell the full story of climate impact.

<h3 style="color: blue;">Section 2 : Disaster Impact Relation to Income Bucket</h3>

Understanding Disaster Impact Across Income Levels
\
\
This section highlights the correlation between disaster impact and income levels. Specifically, we use the inequality ratio = disaster_deaths_missing_per_100k/co2_tons_per_capita to distinguish a nation’s contribution to climate change and the human price it pays for it. We expect to observe that wealth, alongside being a cause of emissions, also serves as the primary shield against mortality.
\
\
By drawing a connection between the inequality ratio and income groups, we aim to extend the correlation between income levels and CO2 emissions to the impact of disasters of wealthier nations.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import plotly.express as px
```

```python
# Create Income Buckets based on GDP
df['income_group'] = pd.qcut(df['gdp_per_capita_ppp_const'], 4, 
                             labels=['Low Income', 'Lower-Middle', 'Upper-Middle', 'High Income'])

# Calculate the Inequality Ratio
df['inequality_ratio'] = df['disaster_deaths_missing_per_100k'] / (df['co2_tons_per_capita'] + 1e-6)

plt.figure(figsize=(12, 7))
# We filter for deaths > 0 to make the log scale meaningful
plot_df = df[df['disaster_deaths_missing_per_100k'] > 0]

sns.scatterplot(
    data=plot_df,
    x='gdp_per_capita_ppp_const',
    y='disaster_deaths_missing_per_100k',
    hue='owid_region',
    alpha=0.6,
    palette='viridis'
)

sns.regplot(
    data=plot_df, 
    x='gdp_per_capita_ppp_const', 
    y='disaster_deaths_missing_per_100k', 
    scatter=False, 
    color='red', 
    logx=True
)

plt.xscale('log')
plt.yscale('log')
plt.title('The Resilience Gap: Wealth vs. Disaster Mortality (2005-2023)', fontsize=15)
plt.xlabel('GDP per Capita (PPP, Constant USD) - Log Scale', fontsize=12)
plt.ylabel('Disaster Deaths per 100k - Log Scale', fontsize=12)
plt.grid(True, which="both", ls="-", alpha=0.2)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', title='Region')
```

#### Observations from the Disaster Deaths-GDP Relationship Scatterplot

The scatter plot above reveals to us a weak, positive relationship between GDP per capita and Disaster Deaths per 100k, specifically between 2005-2023. While the trend line suggests a slight upward slope, the data points are largely scattered, reflecting the unpredictable nature of disaster impacts and the possibility of potential bias due to low death report rates in certain areas.

Let's consider several key patterns that emerge:

**Strong Regional Clustering:** Nations show distinct clustering by geography and income. Europe is heavily concentrated in the lower-right of the graph, indicating a high GDP per capita and low disaster deaths. Conversely, African nations cluster toward the left (lower GDP), yet they exhibit a wide range in mortality, with death rates spanning from $10^{-1}$ to $10^2$ per 100k.

**The "Shield" of Wealth:** Most other regions, including parts of Asia and the Americas, are concentrated near the middle of the distribution. However, the absence of high-income nations in the top-most mortality brackets suggests that while wealth does not eliminate disaster risk, it often prevents the most extreme mass-fatality events.

**Impact of Stochastic Events:** The high degree of scatter demonstrates that disaster mortality is not purely determined by wealth. Significant outliers exist where nations with moderate-to-high wealth still experience high mortality, likely due to specific, high-intensity catastrophic events that overwhelm local infrastructure. Moreover, we also see a potential for low-income nations having incorrect death reportings, highlighted by the large scatter of disaster deaths for African nations.

This relationship is a crucial pillar for our hypothesis. Even though the correlation is weak, the fact that wealthy nations (which we found to be the highest emitters in Section 1) are shielded from the highest death rates, while lower income nations (low emitters) are exposed to the full spectrum of mortality, highlights a profound climate inequality.

```python
map_df = df.groupby(['entity', 'code'])['inequality_ratio'].median().reset_index()

# 2. Get the income group for each country 
# (We take the 'last' one recorded to reflect their most recent status)
income_mapping = df.sort_values('year').groupby('code')['income_group'].last().astype(str).reset_index()

# 3. Merge income back into our summary
map_df_with_income = map_df.merge(income_mapping, on='code')

# 4. Filter for the Top 20
top_20_inequality = map_df_with_income.sort_values(by='inequality_ratio', ascending=False).head(20)

# 5. Plotting
plt.figure(figsize=(12, 10))
sns.barplot(
    data=top_20_inequality, 
    y='entity', 
    x='inequality_ratio', 
    hue='income_group', 
    dodge=False,
    palette='YlOrRd',
    hue_order=['Low Income', 'Lower-Middle', 'Upper-Middle', 'High Income']
)

plt.title('Climate Injustice: Top 20 Nations by Inequality Ratio', fontsize=16)
plt.xlabel('Inequality Ratio (Impact per Unit of Emission)', fontsize=12)
plt.ylabel('Country', fontsize=12)
plt.legend(title='Income Level', loc='lower right')
plt.grid(axis='x', linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```

#### Observations from the Income_Group - Inequality Relationship Bar Graph

The bar graph above shows a strong positive relationship between wealth and the inequality ratio (disaster_deaths_missing_per_100k / co2_tons_per_capita), specifically between 2005-2023.

Let's consider several key patterns that emerge:

**Lower_Income to High_Income:** 15 of the top 20 nations with the greatest inequality ratio (75%) are of low or lower-middle income groups. This underscores a "double burden" where those with the lowest historical contribution to emissions suffer the most severe human consequences during disasters per ton of CO₂ emitted.

**Extreme Outliers:** The top two nations on the list, Malawi (Low Income) and Bhutan (Upper-Middle Income), and number 5 in the list, Lithuania (High Income), exhibit significantly higher inequality ratios than the rest of the top 20.  This indicates localized crises where disaster mortality is exceptionally high relative to their carbon footprints. However, Malawi's high inequality ratio is inferenced to be a consequence of wealth inequality alongside a localized issue based on the trend of a majority of the remaining top 20 nations being low/lower-middle income.

**Regional Concentration:** The visual prevalence of nations from Africa, such as Burundi, Guinea, Burkina Faso, and Liberia, emphasizes a geographic clustering of climate injustice, where entire regions pay a high human price for global atmospheric changes they did not primarily cause.

#### Observations from the Income-Emissions-Disaster_Death-Inequality Ratio Table

The table depicts a relationship among the four variables we generated from the two datasets.

Let's consider several key patterns that emerge:

**Responsibility vs. Impact Divergence:** There is a positive correlation between income and CO2 emissions (Responsibility). High Income nations emit an average of 11.621 tons per capita, over 31 times more than Low Income nations (0.368 tons). Despite this massive difference in contribution to the problem, the human cost is not distributed proportionally.

**Inequality Ratio Gap:** The inequality ratio for Low Income nations (22.399) is nearly 19 times higher than High Income nations (1.205). This shows that for every unit of carbon emitted, lower-income nations suffer almost 20 times the human loss of the wealthiest.

**The Lower-Middle Income "Peak** Interestingly, the Lower-Middle Income group shows the highest absolute average deaths per 100k (9.262). This suggests that nations in the early stages of development face a "vulnerability peak," where they have increased industrial exposure and population density but do not acquired the full "wealth shield" (infrastructure, medical systems, early warning technology) seen in higher income brackets.

Overall, as income increases, the Inequality Ratio significantly decreases (22.39 $\rightarrow$ 7.27 $\rightarrow$ 1.53 $\rightarrow$ 1.20). This confirms that wealth serves as a decoupling mechanism; it allows High Income nations to maintain high emission lifestyles while remaining relatively insulated from the consequences of climate disasters per unit of carbon they produce.

#### Summary: What Section 2 Data Tells Us

Section 2 demonstrates the vulnerability gap. By putting income, disaster impact, and emissions side-by-side it shows structural climate inequality: the nations with the lowest responsibility (Low Income) are paying the highest human price, while the nations with the highest responsibility (High Income) are protected by an economic buffer.
