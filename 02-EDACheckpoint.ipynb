### Load Datasets #1, #2, #3

```python
import pandas as pd

#shared time frame of datasets
START_YEAR = 2005
END_YEAR = 2023

#OWID URLs with short column names 
co2_url = "https://ourworldindata.org/grapher/co-emissions-per-capita.csv?v=1&csvType=full&useColumnShortNames=true"
disasters_url = "https://ourworldindata.org/grapher/deaths-and-missing-persons-due-to-natural-disasters.csv?v=1&csvType=full&useColumnShortNames=true"
gdp_url = "https://ourworldindata.org/grapher/gdp-per-capita-worldbank.csv?v=1&csvType=full&useColumnShortNames=true"

#load dataset, from website, storage option specified
co2 = pd.read_csv(co2_url, storage_options={'User-Agent': 'Our World In Data data fetch/1.0'})
disasters = pd.read_csv(disasters_url, storage_options={'User-Agent': 'Our World In Data data fetch/1.0'})
gdp = pd.read_csv(gdp_url, storage_options={'User-Agent': 'Our World In Data data fetch/1.0'})

print('Loaded datasets:')
print('CO2 shape:', co2.shape)
print('Disasters shape:', disasters.shape)
print('GDP shape:', gdp.shape)
```

### Filter to 2005-2023 and confine to countries

```python
#restrict to shared time frame, will use .copy() for extra safety
co2 = co2[(co2["year"] >= START_YEAR) & (co2["year"] <= END_YEAR)].copy()
disasters = disasters[(disasters["year"] >= START_YEAR) & (disasters["year"] <= END_YEAR)].copy()
gdp = gdp[(gdp["year"] >= START_YEAR) & (gdp["year"] <= END_YEAR)].copy()

#remove aggregate regions such as 'Asia', 'World', etc to only look at countries - keep rows with ISO codes
co2 = co2[co2["code"].notna()].copy()
disasters = disasters[disasters["code"].notna()].copy()
gdp = gdp[gdp["code"].notna()]

print('After filtering:')
print('CO2 Shape:', co2.shape)
print('Disasters shape:', disasters.shape)
print('GDP shape:', gdp.shape)
```

### Inspect and rename key variables for readability

```python
#select and rename key variables after previously analyzing column names with df.columns
co2_clean=co2[['entity', 'code', 'year', 'emissions_total_per_capita']].rename(columns = {'emissions_total_per_capita': 'co2_tons_per_capita'})
disasters_clean = disasters[['entity', 'code', 'year', '_13_1_1__vc_dsr_mtmp']].rename(columns={'_13_1_1__vc_dsr_mtmp': 'disaster_deaths_missing_per_100k'})
gdp_clean = gdp[['entity', 'code', 'year', 'ny_gdp_pcap_pp_kd', 'owid_region']].rename(columns = {'ny_gdp_pcap_pp_kd': 'gdp_per_capita_ppp_const'})

print('CO2 clean columns:', co2_clean.columns.tolist())
print('Disasters clean columns:', disasters_clean.columns.tolist())
print('GDP clean columns:', gdp_clean.columns.tolist())
```

### Merge into single, final wrangled dataframe

```python
#merge using ISO and year
df = co2_clean.merge(disasters_clean[['code', 'year', 'disaster_deaths_missing_per_100k']], on = ['code', 'year'], how = 'inner')
df = df.merge(gdp_clean[['code', 'year', 'gdp_per_capita_ppp_const', 'owid_region']], on = ['code', 'year'], how = 'inner')

#drop any missing values in key variables
df = df.dropna(subset=['co2_tons_per_capita', 'disaster_deaths_missing_per_100k', 'gdp_per_capita_ppp_const']).copy()

#remove duplicate country year rows in case
df = df.drop_duplicates(subset = ['code', 'year'])

print('Final merged dataset shape:', df.shape) 
print('Countries:', df['entity'].nunique())
print('Years:', df['year'].min(), 'to', df['year'].max())
df.head()
```
