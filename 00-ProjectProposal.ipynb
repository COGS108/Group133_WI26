## Research Question

Are higher per-capita $CO_2$ emissions associated with greater climate-related disaster impacts (in the form of higher mortality and reduced GDP growth) for lower-income countries than for higher-income countries?

## Background and Prior Work

Human-driven climate change is largely caused by the accumulation of greenhouse gas (GHG) emissions, namely carbon dioxide ($CO_2$), which accounts for about 66% of human-emitted GHGs. <a name="cite_note-1"></a><sup>1</sup> This results from burning fossil fuels for energy, transportation, manufacturing, and other human-related activities. Global datasets, such as those maintained by Our World in Data, provide long-term measurements of $CO_2$ emissions and their trends across countries, showing clear differences between high- and low-income nations in per-capita emission levels. <a name="cite_note-2"></a><sup>2</sup> 
\
\
The rise in $CO_2$ emissions are not only linked to global warming, but to detrimental climate change, as these shifting climate patterns increase the severity and frequency of extreme weather events such as heatwaves, floods, storms. <a name="cite_note-3"></a><sup>3</sup> These natural disasters carry both large economic and human costs worldwide. In terms of human impacts, international disaster records, such as those by EM-DAT, show that disaster-related mortality rates vary over time and across regions, with certain countries experiencing significantly higher mortality rates per 100,000 people than others. <a name="cite_note-4"></a><sup>4</sup> In addition to the loss of life, these events generate economic damages, including destruction of infrastructure and housing. Research from global monitoring organizations highlight that such economic losses are especially exacerbated in regions with weaker infrastructure and lower capacity for disaster response. <a name="cite_note-5"></a><sup>5</sup>
\
\
There has been growing research and discussion on climate inequality, which explores the ways in which climate change may disproportionately impact different groups of people. Lower-income countries and vulnerable populations typically experience larger relative burdens of climate-driven losses despite contributing far less to greenhouse gas emissions compared to wealthier nations and more privileged populations. <a name="cite_note-6"></a><sup>6</sup> Multiple analyses and systematic reviews show that inequalities in capacity and access to resources can compound the harmful effects of climate change, thus worsening the outcomes for these disadvantaged groups both within and across countries. <a name="cite_note-7"></a><sup>7</sup>
\
\
Economic research supports this link between climate change and inequality. For example, climate-related harms can greatly reduce GDP growth in poorer regions and slow overall economic development, further contributing to the wider disparity in wealth and well-being between countries. <a name="cite_note-8"></a><sup>8</sup> These findings suggest that combining measures of emissions, disaster impacts, as well as economic indicators like GDP per capita into our analysis can show us different countries' climate contribution versus vulnerability.
\
\
By utilizing publicly available datasets on per-capita $CO_2$ emissions, disaster impact statistics, and GDP per capita, we aim to quantify the relationship between human activity and climate outcomes across low- and high-income countries. This approach builds on prior work in climate change science and economic inequality research while applying data science and analysis practices and methods covered in this course.

## Hypothesis


We hypothesize that lower-income countries experience greater climate-related disaster impacts despite having lower per-capita $CO_2$ emissions than that of higher-income countries. We also expect the association between emissions and disaster impacts to differ when countries are grouped by income level.

## Data

1. **Ideal Dataset**: To answer our researh question, the ideal dataset would be a country-year-level panel dataset that combines measures of human activity contributing to climate change, climate-related impacts, and national wealth. 
   1. Some ideal variables would be country identifier/name, year, $CO_2$ emissions per capita (e.g. metric tons per person), climate-related disaster impact (e.g. deaths and missing persons per 100,000 population), GDP per capita (e.g. USD as a proxy for national income)
   2. Ideally, the dataset would contain several thousand observations, covering most countries worldwide over mulriple decades. This scale and time range would allow for adequate comparisons across income groups, countries, and time.
   3. Ideally, this data would be collected by international organizations using standardized reporting methods and metrics. $CO_2$ emissions data would be compiled from national consumption statistics, disaster impact data from global disaster databases, and GDP data from national accounts collected by a standardized and regulated international organization such as the World Bank.
   4. Ideally, the data would be stored in a clean and tiday, tabular CSV format, where eavh row represents a single country year and each columncorresponds to some ideal variable previously defined. Datasets would be merged using country identifiers and years.
2. **Real Datasets**
   1. $CO_2$ Emissions per Capita: https://ourworldindata.org/grapher/co-emissions-per-capita?tab=table : This dataset provides annual, country level measurements of $CO_2$ emissions per person. The primary variable of interest is $CO_2$ emissions per capita, which serves as a proxy for human activity contributing to climate change. The data is publicly available and can be downloaded as a CSV without extra permissions.
   2. Deaths and Missing Persons Due to Natural Disasters: https://ourworldindata.org/grapher/deaths-and-missing-persons-due-to-natural-disasters : This dataset contains country year estimates of deaths and missing persons due to natural disasters, reported as rates per 100,000 population. This variable is used as a proxy for climate related harm. This dataset is derived from the global disaster database, EM-DAT, is publicly available and can be download as a CSV without extra permissions.
   3. GDP per Capita (World Bank): https://ourworldindata.org/grapher/gdp-per-capita-worldbank : This dataset provides annual GDP per capita values in constant USD. GDP per capita is used as a proxy for national income and allows countries to be grouped into lower and higher income categories. The data is publicly available and can be download as a CSV without extra permissions.
